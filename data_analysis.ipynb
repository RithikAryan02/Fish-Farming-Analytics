```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

n = 1000

fish_types = ['Rohu', 'Catla', 'Tilapia']
seasons = ['Summer', 'Monsoon', 'Winter']

data = {
    'Fish_Type': np.random.choice(fish_types, n),
    'Pond_Size': np.random.uniform(0.5, 5, n),
    'Feed_Cost': np.random.randint(10000, 50000, n),
    'Labor_Cost': np.random.randint(5000, 20000, n),
    'Temperature': np.random.uniform(20, 35, n),
    'Rainfall': np.random.uniform(50, 300, n),
    'Season': np.random.choice(seasons, n),
    'Market_Price_per_kg': np.random.randint(100, 250, n),
    'Yield_kg': np.random.randint(500, 3000, n)
}

df = pd.DataFrame(data)

df['Total_Cost'] = df['Feed_Cost'] + df['Labor_Cost']
df['Revenue'] = df['Market_Price_per_kg'] * df['Yield_kg']
df['Profit'] = df['Revenue'] - df['Total_Cost']

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error
```

```python
le = LabelEncoder()

df['Fish_Type'] = le.fit_transform(df['Fish_Type'])
df['Season'] = le.fit_transform(df['Season'])
```

```python
X = df.drop('Profit', axis=1)
y = df['Profit']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)

print("MAE:", mean_absolute_error(y_test, pred))
```
